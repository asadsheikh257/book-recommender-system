## **Popularity Based Recommender System**

```python
merged_df = rating_df.merge(book_df, on='ISBN')
```

```python
merged_df.shape
```

```python
gg = merged_df.groupby('Book-Title').count()['Book-Rating'].reset_index()
gg.rename(columns={'Book-Rating':'num-rating'}, inplace=True)
gg
```

```python
mean_ratings = merged_df.groupby('Book-Title')['Book-Rating'].mean().reset_index()
mean_ratings.rename(columns={'Book-Rating':'avg-rating'}, inplace=True)
mean_ratings
```

```python
f_df = gg.merge(mean_ratings, on='Book-Title')
f_df
```

```python
backup = book_df.merge(f_df, on='Book-Title')
backup.head(1)
```

```python
final_df=f_df[f_df['num-rating'] >= 250]
```

```python
final_df = final_df.sort_values(ascending=False, by='avg-rating')
final_df = final_df.head(50)
```

```python
final_df = final_df.merge(book_df, on='Book-Title').drop_duplicates('Book-Title')
```

```python
popular_df = final_df[['Book-Title', 'Book-Author', 'Year-Of-Publication', 'Image-URL-L', 'num-rating', 'avg-rating']]
```

```python
popular_df.head(1)
```

## **Collabrative Filtering Based Recommender System**

```python
merged_df = rating_df.merge(book_df, on='ISBN')
```

```python
x =  merged_df.groupby('User-ID').count()['Book-Rating'] >= 200
padhe_likhe_users =  x[x].index
```

```python
filtering_rating =  merged_df[merged_df['User-ID'].isin(padhe_likhe_users)]
```

```python
y =  filtering_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books =  y[y].index
```

```python
final_rating =  filtering_rating[filtering_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_rating.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score =  cosine_similarity(pt)
```

```python
def recommend(book_name):
    index =  np.where(pt.index == book_name)[0][0]
    similar_item = sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True)[1:6]
    data = []
    for i in similar_item:
        item = []
        temp_df = book_df[book_df['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-L'].values))

        data.append(item)
    return data
```

```python
recommend('The Fourth Hand')
```
